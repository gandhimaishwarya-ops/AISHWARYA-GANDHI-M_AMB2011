```python
import pandas as pd
production=pd.Series(
    [30,60,45,25],
    index=["Wheat","Rice","Maize","Barley"]
)
print(production)
```

```python
#loc=location by label or value
print(production.loc["Rice"])

print(production.loc["Wheat"])
```

```python
#iloc= location by index
print(production.iloc[0])
```

```python
print(production.sum())
print(production.mean())
print(production.max())#Maximum production crop
print(production.idxmax())   #it print the index label of the maximum value in the series

print(production.idxmax(),"is having maximum value of ", production.max())
```

```python
#increased all crop production by 10%(good harvest year)
increased_production=production*1.10
print(increased_production)
```

```python
#Crop with production grater than 40 tons
high_production=production[production>40]
print(high_production)
print()

print(production>40)
```

```python
import pandas as pd
data={
    "Farm":["Farm A","Farm B","Farm C","Farm D"],
    "Crop":["Wheat","Rice","Wheat","Maize"],
    "Area_hectares":[10,15,8,12],
    "Production_tons":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
#yield=production/area
df["Yield_tons_per_hectare"]=df["Production_tons"]/df["Area_hectares"]
print(df)
```

```python
#Filter data
#Farms growing wheat
print(df[df["Crop"]=="Wheat"])
print()
wheat_farms=df[df["Crop"]=="Wheat"]
print(wheat_farms)

```

```python
#farm with yeild greater than 3
high_yield=df[df["Yield_tons_per_hectare"]>3]
print(high_yield)
```

```python
#sort the farms by highest production
sorted_df=df.sort_values(by="Production_tons",ascending=False)
print(sorted_df)
```

```python
#Group By operation -Average yeild per crop
avg_yield=df.groupby("Crop")["Yield_tons_per_hectare"].sum()
print(avg_yield	)
```
