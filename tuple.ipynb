1.3.Tuple iteration using while loop

```python
T=("Pratham",'Sharma',3.14,3)
print(T)
i=0
print("")
while i<len(T):
	print(T[i])
	i=i+1
```

1.4. Tuple iteration using for loop

```python
T=("Pratham",'Sharma',3.14,3)
print(T)                                                 #reffering to the indexes
print("")
for i in range(len(T)):
	print(T[i])
```
