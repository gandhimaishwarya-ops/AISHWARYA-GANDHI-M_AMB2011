```python
crops=0
points=0

print("Welcome to farm game")

print("1.plant crop")
print("2.water plants")
print("3.harvest crop")
print("4.exit")
while True:
    ch=int(input("enter your choice"))
    if ch==1:
      crops=crops+1
      print("one crop planted")
    elif ch==2:
      if crops>0:
        print("Crops waterd")
      else:
        print("No crops to water")
    elif ch==3:
      if crops>0:
        points+=crops*10
        print("Crops harvested")
      else:
        print("No crops to harvest")
    elif ch==4:
      print("Game over Total points-",points)
      break
    else:
      print("Invalid choice")

```

**FISH GAME**

```python
import random
#Initialize inventory and money
inventory={}
total_money=0

#Fish types and prices
fish_prices={
    "Rohu":12,
    "Catla":18,
    "Tuna":25
    }

print("\n-Fishing Game Menu")
print("1.Catch fish")
print("2.View inventory")
print("3.Sell Fish")
print("4.Exit")

#Game loop

while True:
    choice=int(input("Choose an option:"))
  # catch fish
    if choice==1:
        #Randomly select a fish
        fish=random.choice(list(fish_prices.keys()))
        inventory[fish]=inventory.get(fish,0)+1
        print(f"You caught a {fish}!")
  #View Inventory
    elif choice==2:
        if inventory=={}:
            print("No Fish")
        else:
            print("Your Inventory:")
            for fish,count in inventory.items():
                print(f"{fish}:{count}")

  #Sell fish
    elif choice==3:
        if inventory=={}:
            print("No fish to sell")
        else:
            earnings=0
            for fish,count in inventory.items():
                earnings+=fish_prices[fish]*count
            total_money+=earnings #total_money=total_money earnings
            inventory.clear()
            print(f"You sold all fish for ${earnings}") # Corrected an extra parenthesis in the f-string format

  #exit
    elif choice==4:
        print(f"Final money earned:${total_money}")
        print("Thanks for playing!")
        break
    else:
        print("Invalid choice.Please try again.")
```
