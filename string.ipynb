```python
#assign a variable with"Today is 19-Jan-2026 M0NDAY"
#Calculate number of vowels,consonates,blackspaces
#and other character


str="Today is 19-Jan-2026 Monday"
B=V=C=D=O=0
for i in str:
  if i=='':
    B=B+1
  elif i in 'aeiouAEIOU':
    V=V+1
  elif i.isalpha():
    C=C+1
  elif i.isdigit():
    D=D+1
  else:
    O=O+1
print(B)
print(V)
print(C)
print(D)
print(O)

```

**Extracting ALL strings with a prefix**

```python
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
filtered=[]

for crop_code in codes:
  if crop_code.startswith("CRP"):    #true
    filtered.append(crop_code)
print(filtered)
```

**Extact and clean in one go**

```python
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
clean_codes=[]

for code in codes:
  if code.startswith("CRP_"):
    clean_codes.append(code[4:])


print(clean_codes)
```

**split method**

```python
codes=[
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
for crop_code in codes:
  if crop_code.startswith("CRP_"):
    code,crop,year=crop_code.split("_")
    print(crop,year)
```

**Mini Agriculture Data Clearing Example**(*#A consolidated example using multiple string operation*)

```python
raw_record="Rice,1200,Black Soil**"
clean_record=raw_record.strip()
clean_record1=clean_record.rstrip("*")
crop,yield_kg,soil=clean_record1.split(",")
crop=crop.title()

print(f"Crop:{crop}")
print(f"Yield:{yield_kg}kg")
print(f"Soil Type:{soil}")
```
