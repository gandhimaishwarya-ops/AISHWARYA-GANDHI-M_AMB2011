```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure data is loaded
df = pd.read_csv("agri_data.csv")

# Recalculate yearly_production
yearly_production = df.groupby("Year")["Production_Tons"].sum()

# custom styled bar plot
plt.figure(figsize=(2,4))

bars = plt.bar(yearly_production.index,
             yearly_production.values,
             color='lightgreen',
             edgecolor='black')

# add value label on the top bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height,
             str(height),
             ha='center',
             va='bottom')

plt.xlabel("Year")
plt.ylabel("Total Production(Tons)")
plt.title("Yearly Total Production with Values")
plt.show()
```

```python
import pandas as pd
import numpy as np

#Create data with missing values
data=pd.DataFrame({
    'plot':['A','B','C','D','E'],
    'yield_kg':[2500,np.nan,2800,np.nan,2600],
    'rainfall':[150,200,np.nan,180,170],
    'temperature':[28,30,29,np.nan,27]
})

print("==Data with missing values==")
print(data)

#==DETECTNG MISSING DATA==
print("\n==Detecting missing values==")
print("Is null?")
print(data.isnull())
print(f"\nTotal missing per column:")
print(data.isnull().sum())
```

```python
#==FILLING MISSING VALUES==
print("\n==Filling Strategies==")

#Fill with mean
data_mean=data.copy()
print("Original yield_kg:")
print(data_mean['yield_kg'])

# Corrected syntax: assign the result of fillna back to the column
data_mean['yield_kg'] = data_mean['yield_kg'].fillna(data_mean['yield_kg'].mean())

print('\nFill with mean result:')
print(data_mean['yield_kg'])
```

```python
#Fill with median
data_median=data.copy()

print("Original rainfall:")
print(data_median['rainfall'])

# Fixed: changed print_median to data_median
data_median['rainfall']=data_median['rainfall'].fillna(data_median['rainfall'].median())

print("\nRainfall after filling with median:")
print(data_median['rainfall'])
```

```python
#Forward fill (use previous value)
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')
print("\nForward fill:")
print(data_ffill)
#ffill-It replaces mising values(NaN)
# With the last valid(non-missing) value above it in the column
```

```python
#===DROOPING MISSING DATA===
print("\n==Dropping Rows===")
data_drooped=data.dropna()
print(data_drooped)
print(f"Original rows:{len(data)},After drop:{len(data_drooped)}")
```
