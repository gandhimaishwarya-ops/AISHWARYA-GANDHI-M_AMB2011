```python
import pandas as pd
dataset=pd.read_csv('Salary_Data.csv')
print(dataset)
```

```python
X=dataset.iloc[:,0:1].values
Y=dataset.iloc[:,-1].values
print(X)
print(Y)
```

```python
#Splitting the dataset into training set and testing set
from sklearn.model_selection import train_test_split
# Ensuring X and Y are defined and splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=1/3, random_state=0)

print("Data split successfully.")
print("X_test shape:", X_test.shape)
```

```python
import numpy as np

#Calculate the mean values of input and output
X_mean=np.mean(X_train)
y_mean=np.mean(y_train)
print(X_mean)
print(y_mean)
```

```python
#Calculate intercept and coefficient
n=len(X_train)
num,den=0,0

for i in range(n):
    num+=(X_train[i]-X_mean)*(y_train[i]-y_mean)
    den+=(X_train[i]-X_mean)**2

b1=num/den
b0=y_mean-(b1*X_mean)
print("b1 (slope):", b1)
print("b0 (intercept):", b0)
```

```python
#predicting results on test dataset
y_pred_code=b0+b1*X_test
print(y_pred_code)
```

```python
print(y_test)
```

```python
#mean square error
n1=len(y_test)
mse=0
for i in range(n1):
    mse+=(y_test[i]-y_pred_code[i])**2

mse=mse/n1
print(mse)
rmse=np.sqrt(mse)
print(rmse)
```

**machine learning alogorithm**

```python
#create an object of the alogorithm / model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
print(y_pred)
```

```python
print('Intercept',reg.intercept_)
print('coefficient',reg.coef_)
```

```python
#Evaluation metrics
from sklearn import metrics
print("Mean absolute error:")
print(metrics.mean_absolute_error(y_test,y_pred))
```

```python
print("Mean squared error:", metrics.mean_squared_error(y_test, y_pred))
```

```python
print("Root mrsn square error:",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
import matplotlib.pyplot as plt

#Visualisation
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,reg.predict(X_train),color='blue')
plt.title('Salary vs Experience-Training set')
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.show()
```
