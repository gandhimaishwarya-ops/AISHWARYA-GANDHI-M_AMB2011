```python
#Scope of variables(local variable)
#with parameter

def fun(x):
  print("Local x is:",x)
  x=2
  print("Changed local x to:",x)
x=50
fun(x)
print(" x is still:",x)
```

```python
#global variable
def  fun():
  global y
  print("Y is:",y)
  y=2
  print("local y to:",y)
y=50
fun()
print("value of y is:",y)
```

```python
total=0
#function definition is here
def sum(arg1,arg2):
  #add both the parameters and return them.
  total=arg1+arg2 #here total is local variable
  print("Inside the function local total:",total)
  return total

#now can you call sum function
sum(10,20)
print("Outside the function global total:",total)
```

***Types of function or Formal Arguments***

**1.REQUIRED ARGUMENT**

```python
def printme(str):
  print(str)
  return
#printme() #statement gives out the error
#printme("hello","bye") #statement gives out the error
printme("Hello")
printme("12345")
```

**2.Keyword argument or named argument**

```python
#keyword argument
def printme(str):
  #this prints a passed string into this function
  print(str)
  return
printme("Hello") #required argument
#now can you call printme function
printme(str="My string")#named argument

#printme(abc="My string")  #error will occur
#abc is unknown argument
```

**3.Default Argument**

```python
def func(a,b=5,c=10):
  print("a is",a,"and b is",b,"and c is",c)
func(3,7) #a=3 b=7 c=10  a-required  b-keyword  c=default
func(25,c=24) #a=25,b=5,c=24  a-required  b-default  c-keyword
func(c=50,a=100) #parameter can be in any order
```

```python
def func(x,y=5,z=10):
  print("x is",x,"and y is",y,"and z is",z)
func(4,5,6)
func(2,z=9)
func(2,3)
func(4,5)
```

```python
def func(a,b,c):
  print("a is",a,"and b is",b,"and c is",c)
func(3,7) #a=3 b=7 c=10  a-required  b-keyword  c=default
```

**4.Variable-length Arguments**

```python
def check(name,*marks):
  print(name)
  print(marks)
check("XYZ",10,20,30)

NOTE:#Default argument should be after non default argument
#def.check(age=10,name)--------error
#def.check(name,age=10)--------correct way
```
